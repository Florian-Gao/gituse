```python
#二叉搜索树 binary search tree

class BiTreeNode:
    def __init__(self,data):
        self.data=data
        self.lchild=None
        self.rchild=None
        self.parent=None
        
class BST:
    def __init__(self,li):
        self.root=None
        if li:
            for i in li:
                self.insert(i)
                
    def insert(self,key):
        if not self.root:
            self.root=BiTreeNode(key)
        else:
            p=self.root
            while p:
                if key<p.data:
                    if p.lchild: #如果左子树有东西，继续往下走
                        p=p.lchild
                    else:  #如果没有，将key添加为左子树
                        p.lchild=BiTreeNode(key)
                        p.lchild.parent=p
                        break
                elif key>p.data:
                    if p.rchild: #如果右子树有东西，继续往下走
                        p=p.rchild
                    else:  #如果没有，将key添加为右子树
                        p.rchild=BiTreeNode(key)
                        p.rchild.parent=p
                        break
                else:
                    break
                    
    def query(self,k):#查看值是不是在树内
        p=self.root
        while p:
            if k==p.data:
                return True
            elif k<p.data:
                p=p.lchild
            else: 
                p=p.rchild
        return False
                     
    def traverse(self): #输出树
        def mid_order(root):
            if root:
                mid_order(root.lchild)
                print(root.data)
                mid_order(root.rchild)
        mid_order(self.root)
        
    
    
    
    #删除操作
    def _remove_node_1(self,node):
            if not node.parent: #如果这个节点没有母亲，则是根结点
                self.root=None #直接删掉，就是None
            elif node==node.parent.lchild:
                node.parent.lchild=None
            else:
                node.parent.rchild=None

    def _remove_node_21(self,node): #如果这个节点只有右孩子
        if not node.parent:  #如果要删除根结点
            self.root=node.rchild #把右半边上挪
            self.root.parent=None #再把之前的根结点删除
        elif node==node.parent.lchild:# 如果这个节点是其母节点的左孩子
            node.parent.lchild=node.rchild
            node.rchild.parent=node.parent
        else:
            node.parent.rchild=node.rchild
            node.rchild.parent=node.parent

    def _remove_node_22(self,node):
        if not node.parent:
            self.root=node.lchild
            self.root.parent=None
        elif node==node.parent.lchild:
            node.parent.lchild=node.lchild
            node.lchild.parent=node.parent
        else:
            node.parent.rchild=node.lchild
            node.lchild.parent=node.parent

    def delete(self,val):
        if self.root:
            node=self.query(val)  #先确定这个点的位置
            if not node: #如果不在
                return False
            if not node.lchild and not node.rchild: #叶节点
                self._remove_node_1(node)
            elif not node.lchild: #2.1只有右孩子
                self._remove_node_21
            elif not node.rchild: # 2.2 只有左孩子
                self._remove_node_22
            else: #3.两个孩子
                #找到右子树的最小点或左子树的最大点
                min_node=node.rchild
                while min_node.lchild:
                    min_node=min_node.lchild
                node.data=min_node.data
                #删除该节点
                if min_node.rchild:
                    self._remove_node_21(min_node)
                else:
                    self._remove_node_1(min_node)
                    
li=[5,1,2,3,6,9,4,8,7]
tree=BST(li)
tree.traverse()
tree.query(5)
tree.delete(1)
```

```python
#链表
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkList:
    def __init__(self,li,method='head'):
        self.head=None
        self.tail=None
        if method=='head':
            self.create_linklist_head(li)
        if method=='tail':
            self.create_linklist_tail(li)
            

    def create_linklist_head(self,li):
        self.head=Node(0)
        for i in li:
            n=Node(i)
            n.next=self.head.next
            self.head.next=n
            self.head.data+=1
        
    def create_linklist_tail(self,li):
        self.head=Node(0)
        self.tail=self.head
        for i in li:
            n=Node(i)
            self.tail.next=n
            self.tail=n
            self.head.data+=1
            
    def traverse(self):
        p=self.head.next
        while p:
            print(p.data)
            p=p.next
            
    def __len__(self):
        return self.head.data
    
li=[1,2,3,4,5]
ll=LinkList(li,method='tail')
ll.traverse()
len(ll)
```
